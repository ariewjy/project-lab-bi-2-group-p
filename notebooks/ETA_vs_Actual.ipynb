```python
orders_data = pd.read_csv('olist_orders_dataset.csv')
orders_data
```

```python
persentase_null_values = orders_data.isnull().sum()/len(orders_data)*100
persentase_null_values
```

```python
orders_data.dtypes
```

```python
count_order_status = orders_data['order_status'].value_counts()
count_order_status
```

```python
orders_data['order_purchase_timestamp'] = pd.to_datetime(orders_data['order_purchase_timestamp'], errors='coerce')
orders_data['order_approved_at'] = pd.to_datetime(orders_data['order_approved_at'], errors='coerce')
orders_data['order_delivered_carrier_date'] = pd.to_datetime(orders_data['order_delivered_carrier_date'], errors='coerce')
orders_data['order_delivered_customer_date'] = pd.to_datetime(orders_data['order_delivered_customer_date'], errors='coerce')
orders_data['order_estimated_delivery_date'] = pd.to_datetime(orders_data['order_estimated_delivery_date'], errors='coerce')
```

```python
orders_data.info()
```

```python
orders_data['order_approved_at_day'] = pd.to_datetime(orders_data['order_approved_at']).dt.to_period('D')
orders_data['order_delivered_customer_date_day'] = pd.to_datetime(orders_data['order_delivered_customer_date']).dt.to_period('D')
orders_data['order_estimated_delivery_date_day'] = pd.to_datetime(orders_data['order_estimated_delivery_date']).dt.to_period('D')
```

```python
orders_data
```

```python
orders_data['ETA'] = (orders_data['order_estimated_delivery_date_day'] - orders_data['order_approved_at_day']) / np.timedelta64(1, 'D')
orders_data['ETA']
```

```python
orders_data['Actual'] = (orders_data['order_delivered_customer_date_day'] - orders_data['order_approved_at_day']) / np.timedelta64(1, 'D')
orders_data['Actual']
```

```python
ETA_vs_Actual = orders_data[['ETA', 'Actual']]
ETA_vs_Actual
```

```python
ETA_vs_Actual.to_csv(r'ETA_vs_Actual.csv')
```
